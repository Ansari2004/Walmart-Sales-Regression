```python
import pandas as pd
import numpy as np
import os
```

```python
# Load the datasets
features = pd.read_csv('../data/features.csv')
stores = pd.read_csv('../data/stores.csv')
train = pd.read_csv('../data/train.csv')
```

```python
# merging the datasets (store and features with the train and test datasets)

train = train.merge(stores, on='Store', how='left')
train = train.merge(features, on=['Store', 'Date', 'IsHoliday'], how='left')
```

```python
# filling missing markdowns with 0 fill

markdown_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
train[markdown_cols] = train[markdown_cols].fillna(0)
```

```python
# converting date to datetime and sorting by store and date to fill missing values in CPI and Unemployment columns

train['Date'] = pd.to_datetime(train['Date'])
train = train.sort_values(['Store', 'Date'])

# using forward fill to fill missing values in CPI and Unemployment columns for each store
train[['CPI', 'Unemployment']] = (
    train.groupby('Store')[['CPI', 'Unemployment']].ffill()
)

train['IsHoliday'] = train['IsHoliday'].astype(int)
```

```python
# extracting date features (year, month, week) from the date column

train['Year'] = train['Date'].dt.year
train['Month'] = train['Date'].dt.month
train['Week'] = train['Date'].dt.isocalendar().week.astype(int)
```

```python
# Prepare the final training dataset

# Create folder if it doesn't exist
os.makedirs('../data', exist_ok=True)

# Prepare final dataset
final_train = train.copy()
final_train = final_train.sort_values(['Store', 'Date'])

# Save inside data folder
final_train.to_csv('../data/final_train.csv', index=False)

print("final_train.csv saved in data/ folder successfully!")
```
